```python
# WI_Bomdila village
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- 1. CONFIGURATION ---
output_folder = r"\Himalaya_1"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# HIGH READABILITY SETTINGS (Optimized for A4 at 30cm)
plt.rcParams.update({
    'font.size': 13,            # Base font size increased
    'axes.labelsize': 11,       # Axis labels
    'axes.titlesize': 12,       # Plot titles
    'legend.fontsize': 9,      # Legend text
    'xtick.labelsize': 9,      # Tick labels
    'ytick.labelsize': 9,
    'lines.linewidth': 1.5,     # Significantly thicker lines
    'lines.markersize': 5,      # Larger markers for quick identification
    'axes.linewidth': 1,      # Thicker axis spines
    'grid.linewidth': 0.5,      # Visible but not distracting grid
    'font.family': 'sans-serif' # Cleaner for technical reading
})

# --- 2. DATA & CALCULATIONS ---
data = {
    'Depth': [0.25, 0.5, 0.75, 1, 2, 3, 4, 5, 6.0], 
    'SiO2': [66.89, 67.12, 63.39, 65.06, 61.23, 64.05, 61.86, 62.34, 73.34],
    'Al2O3': [17.68, 17.7, 14.82, 16.63, 10.21, 11.24, 10.63, 11.92, 14.26],
    'MgO': [2.36, 1.29, 0.72, 1.42, 0.85, 0.61, 0.81, 0.63, 0.73],
    'CaO': [0.46, 0.69, 0.85, 0.67, 0.64, 0.73, 0.75, 0.62, 0.97],
    'Na2O': [0.29, 0.85, 1.44, 1.03, 1.88, 1.9, 1.57, 1.14, 2.64],
    'K2O': [2.04, 2.98, 4.22, 1.66, 5.34, 5.66, 6.28, 5.12, 4.54],
    'TiO2': [1.64, 0.82, 0.45, 0.89, 0.44, 0.45, 0.51, 0.54, 0.37]
}
df = pd.DataFrame(data)
bedrock_depth = df['Depth'].max()

df['R'] = df['SiO2'] / df['Al2O3']
df['WIP'] = 100 * ((2*df['Na2O']/0.35) + (df['MgO']/0.9) + (2*df['K2O']/0.25) + (df['CaO']/0.7))
df['V'] = (df['Al2O3'] + df['K2O']) / (df['MgO'] + df['CaO'] + df['Na2O'])
df['CIA'] = 100 * (df['Al2O3'] / (df['Al2O3'] + df['CaO'] + df['Na2O'] + df['K2O']))
df['CIW'] = 100 * (df['Al2O3'] / (df['Al2O3'] + df['CaO'] + df['Na2O']))
df['PIA'] = 100 * ((df['Al2O3'] - df['K2O']) / (df['Al2O3'] + df['CaO'] + df['Na2O'] - df['K2O']))
s_t, s_a, a_t = df['SiO2']/df['TiO2'], df['SiO2']/df['Al2O3'], df['Al2O3']/df['TiO2']
df['STI'] = 100 * (s_t / (s_t + s_a + a_t))

# --- 3. PLOTTING LOGIC ---
indices_config = {
    'R': {'c': '#E6194B', 'm': 'o'}, 'WIP': {'c': '#3CB44B', 'm': 's'},
    'V': {'c': '#4363D8', 'm': '^'}, 'STI': {'c': '#F58231', 'm': 'D'},
    'CIA': {'c': '#911EB4', 'm': 'v'}, 'CIW': {'c': '#42D4F4', 'm': '*'},
    'PIA': {'c': '#f032e6', 'm': 'p'}
}

group1 = ['R','V']
group2 = ['CIA', 'CIW', 'PIA', 'STI']
group3 = ['WIP']
all_indices = list(indices_config.keys())

def save_styled_plot(filename, subset, is_log=False):
    fig, ax = plt.subplots(figsize=(2.5, 5))
    for idx in subset:
        ax.plot(df[idx], df['Depth'], marker=indices_config[idx]['m'], 
                color=indices_config[idx]['c'], label=idx)
    
    if is_log: ax.set_xscale('log')
    
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    ax.set_xlabel("Weathering Index Value" + ("\n(Log Scale)" if is_log else ""), fontweight='bold', labelpad=15)
    plt.title('Depth-alteration Profile (Log Scale)' if is_log else 'Depth-alteration Profile', fontsize=14, fontweight='bold', pad=60)
    ax.set_ylabel("Depth (m)", fontweight='bold')
    ax.axhline(y=bedrock_depth, color='black', linestyle='--', linewidth=2, label='Bedrock')
    ax.grid(True, which='both', linestyle='-', alpha=0.8)
    ax.legend(loc='center left', bbox_to_anchor=(1.5, 0.5), frameon=True)
    
    plt.savefig(os.path.join(output_folder, filename), dpi=300, bbox_inches='tight')
    plt.close()

# --- 4. EXECUTION ---
save_styled_plot("Group_1.svg", group1)
save_styled_plot("Group_2.svg", group2)
save_styled_plot("Group_3.svg", group3)
save_styled_plot("Combined_Indices.svg", all_indices)
save_styled_plot("Combined_Log_Profile.svg", all_indices, is_log=True)

print(f"Success! The 4 requested charts have been saved to: {output_folder}")
```
